```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from util import generate_scale_free_series_parallel_graph
```

```python
graph = generate_scale_free_series_parallel_graph(100000, 2.3)

degree_dist = graph.degrees
```

```python
# Plot the degree distribution in log-log scale
plt.loglog(degree_dist.keys(), degree_dist.values())
sns.scatterplot(x=list(degree_dist.keys()), y=list(degree_dist.values()))

plt.ylim(1, 10000)
plt.xscale('log')
plt.yscale('log')

# Label the axes
plt.xlabel("Degree")
plt.ylabel("Number of nodes")
plt.title("Log-log plot of scale-free series-parallel graph ($\gamma = 2.3$)")
# Show the plot
plt.show()
```
