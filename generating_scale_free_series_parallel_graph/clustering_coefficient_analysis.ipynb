```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('result.csv')
```

```python
df.info()
```

```python
sns.lineplot(data=df, x='num_of_nodes', y = 'clustering_coefficient', hue='gamma', markers=True, style='gamma', palette='tab10')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Clustering Coefficient vs Number of Nodes for different $\gamma$ in Scale-Free SPGs')
plt.xlabel('Number of Nodes')
plt.ylabel('Clustering Coefficient')
```
