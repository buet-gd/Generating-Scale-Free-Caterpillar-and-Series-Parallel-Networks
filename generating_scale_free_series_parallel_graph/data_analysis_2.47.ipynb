```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('result_2.47.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.histplot(df['num_of_nodes'], kde=False, bins=20)
```

```python
sns.set_theme(style="whitegrid")
sns.regplot(x="num_of_nodes", y="average_shortest_path_length", data=df)
# plt.yscale('log')
plt.show()
```

```python
sns.displot(df, x="clustering_coefficient", kde=True)
plt.show()

sns.set_theme(style="whitegrid")
sns.scatterplot(x="num_of_nodes", y="clustering_coefficient", data=df)
plt.show()

sns.set_theme(style="whitegrid")
sns.regplot(x="num_of_nodes", y="clustering_coefficient", data=df)
plt.show()
```

```python
df.clustering_coefficient.mean(), df.clustering_coefficient.std()
```

```python
sns.displot(df, x="assortativity", kde=True)
plt.show()

sns.set_theme(style="whitegrid")
sns.scatterplot(x="num_of_nodes", y="assortativity", data=df)
plt.show()

sns.set_theme(style="whitegrid")
sns.regplot(x="num_of_nodes", y="assortativity", data=df)
plt.show()
```

```python
sns.displot(df, x="avg_degree_centrality", kde=True)
plt.show()

sns.set_theme(style="whitegrid")
sns.scatterplot(x="num_of_nodes", y="avg_degree_centrality", data=df)
plt.show()
```

```python
sns.displot(df, x="avg_closeness_centrality", kde=True)
plt.show()

sns.set_theme(style="whitegrid")
sns.scatterplot(x="num_of_nodes", y="avg_closeness_centrality", data=df)
plt.show()
```

```python
sns.displot(df, x="avg_betweenness_centrality", kde=True)
plt.show()

sns.set_theme(style="whitegrid")
sns.scatterplot(x="num_of_nodes", y="avg_betweenness_centrality", data=df)
plt.show()
```
